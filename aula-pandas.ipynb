```python
import pandas as pd
```

```python
data = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv',encoding='latin1', parse_dates='Year')
data.head()
```

```python
data.info()
```

```python
data['Year'] = pd.to_datetime(data['Year'], format='%Y')
```

```python
data.set_index('Year', inplace=True)
```

```python
data[['Sobrenome', 'Nome']] = data['Athlete'].str.split(',', expand=True, n=1)
data
```

```python
data.describe(include='all')
```

```python
data[data['Gender'] == 'Men'].groupby(by='Year')['Medal'].count()
```

```python
data[(data['Gender'] == 'Men') & (data['Country_Code']=='USA') ].groupby(by='Year')['Medal'].count()
```
