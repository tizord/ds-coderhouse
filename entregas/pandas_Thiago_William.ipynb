## Prática com Pandas e Séries Cronológicas

```python
import pandas as pd
```

A base a ser utilizada nesse exercício é a base "Summer-Olympic-medals-1976-to-2008.csv"

```python
dados = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding='latin1', parse_dates=['Year'])
```

```python
dados.head()
```

### 1. Construa um objeto de série cronológica com o índice igual ao ano dos jogos olímpicos.

```python
dados.set_index('Year', inplace=True)
dados
```

### 2. Separar os nomes próprios de sobrenomes em duas colunas distintas, usando vírgula como separador

```python
dados[['Sobrenome', 'Nome']] = dados['Athlete'].str.split(',', expand=True, n=1)
dados
```

### 3. Obter medidas resumidas do conjunto de dados. Qual país ganhou mais medalhas?

```python
medalhas_por_pais = dados.groupby('Country')['Medal'].count()
print(medalhas_por_pais.sort_values(ascending=False))
dados[['Country', 'Medal']].describe()
```

### 4. Construir uma tabela mostrando quantas medalhas foram ganhas pelos homens no total em cada ano em que o evento foi realizado.

```python
dados[dados['Gender']=='Men'].groupby(by='Year')['Medal'].count()
```
